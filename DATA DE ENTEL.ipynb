```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")
```

```python
driver_path = r"C:\Users\ANZONY\Downloads\chromedriver.exe"
driver = webdriver.Chrome(driver_path , options =options)
driver.get('https://catalogo.movistar.com.pe/')
time.sleep(4)
boton = driver.find_element(By.XPATH,'//*[@id="tipoContrato"]/div[3]/label/div/div').click()
```

```python
i = 1
caja = "vacia"
Nombres = []
Precios= []
cuotas= []
j = 1

```

```python
while caja !=0 :
    
    data=driver.find_elements(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a')
    
    # NOMBRES 
    try :
                j = 1
                nombre = data[0].find_element(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a/div[{j}]/div/h3').text
                Nombres.append(nombre)
                print(nombre)
                
    except NoSuchElementException:
                j = 2
                nombre = data[0].find_element(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a/div[{j}]/div/h3').text
                Nombres.append(nombre)
                print(nombre)

        # Precios
    try :
            j = 1
            precio=data[0].find_element(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a/div[{j}]/div/div[1]/strong').text
            Precios.append(precio)
            print(precio)
    except NoSuchElementException:
            j = 2
            precio=data[0].find_element(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a/div[{j}]/div/div[1]/strong').text
            Precios.append(precio)  

        # Cuotas
    try :
            j = 1
            cuota=data[0].find_element(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a/div[{j}]/div/div[2]/strong').text
            cuotas.append(cuota)
            print(cuota)

    except NoSuchElementException:
            j = 2
            cuota=data[0].find_element(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a/div[{j}]/div/div[2]/strong').text
            cuotas.append(cuota)
            print(cuota)

    i += 1
    print(i)

    # PARA PARAR
    data=driver.find_elements(By.XPATH,f'//*[@id="grid-equipos"]/div[2]/ul/li[{i}]/a')
    caja = len(data)

    
    
    
```

```python
Movistar = { "Nombre" : Nombres ,"Precios": Precios, "Cuotas":cuotas}
```
