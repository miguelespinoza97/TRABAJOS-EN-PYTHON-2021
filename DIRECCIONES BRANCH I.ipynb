```python
# Exporto las librerias
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re
from selenium.webdriver.common.alert import Alert
from selenium.common.exceptions import NoSuchElementException
```

```python
# Desabilitamos las extensiones
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')
```

```python
driver_path = "./chromedriver_win32/chromedriver.exe"

driver = webdriver.Chrome(driver_path, options=options)

# Inicializamos el navegador
driver.get('https://gov4c.kz/en/contacts/adresa/')
```

```python
adress=[]
nombre=[]
branch=[]
```

```python


for j in range (1,19,1) :
    
    boton=driver.find_element_by_xpath(f'/html/body/main/div[2]/div/div[2]/div/div/div[{j}]').click()

    
    caja="nada"

    i =3
    
    
    sleep(3)

    while caja!=0:
    
        try:
        #Direcciones
        
            adre=driver.find_element_by_xpath(f'/html/body/main/div[2]/div/div[2]/div/div/article/table/tbody/tr[{i}]/td[5]').text    
            adress.append(adre)
            i=i+1
            print(adre)
        
        #Nombre de branch
            nom_1= driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/h3').text
            branch.append(nom_1)
            nom= driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/h3').text
            match_0=re.findall(r"(?<=in the city of ).*",nom)
            match_1=re.findall(r"(?<= in the city ).*(?=)",nom)
            match_2=re.findall(r"(?<= in the ).*(?= )",nom)
            match_3=re.findall(r"(?<=in the).*(?= )",nom)
            match_4=re.findall(r"(?<=in ).*(?= )",nom)
            
            if match_0:
                dire=match_0[0]
                
            elif match_1:
                dire=match_1[0]
                
            elif match_2:
                dire=match_2[0]
                
            elif match_3:
                dire=match_3[0]
                
            elif match_4:
                dire=match_4[0]
                
            else:
                dire=nom
                
            print(dire)
            nombre.append(dire)
    
        except:
            
            caja=0
            print("acabo")
    
    sleep(3)
    
    driver.back()
    




                               
```

```python
dicc={"Adress":adress,"Branch":branch,"Nombre":nombre}
```

```python
registro=pd.DataFrame(dicc)
registro
```
