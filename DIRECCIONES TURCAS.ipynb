```python
# Exporto las librerias
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re
from selenium.webdriver.common.alert import Alert
from selenium.common.exceptions import NoSuchElementException
```

```python
# Desabilitamos las extensiones
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
driver_path = "./chromedriver_win32/chromedriver.exe"

driver = webdriver.Chrome(driver_path, options=options)

# Inicializamos el navegador
driver.get('https://www.nvi.gov.tr/il-mudurlukleri')

#for item in items:
#     dos = item.get_attribute("href")
#     link.append(dos)
#     print (dos)
```

```python

# Listas nombres estaran contenidas los link y nombres
LISTA_URLS=[]
nombres=[]


for j in range (0,21,1):
    
    caja="no hay"
    i=1

    while caja != 0:

        driver.find_elements_by_xpath('//*[@id="w8852f9c8040342cf8a01134153d98feb"]/div/div/div/div[1]/div[2]')

        try:
            items = driver.find_element_by_xpath(f'//*[@id="{j}"]/div[2]/ul/li[{i}]/a')
            LISTA_URLS.append( items.get_attribute("href") )
            i=i+1
            textos=items.text
            nombres.append(textos)
            print(items.text)
        except:
            caja=0
            print("no hay")
            
```

```python
# lista de latitudes , longitudes , direcciones
datax=[]
datay=[]
region=[]
direccion=[]
```

```python
# Insertamos los link :

for url in LISTA_URLS:
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content,"html.parser")
    texto=soup.find("div", class_="iletisimalani")
    lista=[]
    for x in texto:
        lista.append(str(x))

    if lista:
        lista_2= lista[9]
        #print(lista_2)

    # Data x
    locx_0= re.findall(r"(?<=konumx\: \").*(?=\", konumy)", lista_2)
    locx_1 =(" ".join(locx_0))
    locx_2=(float(locx_1))
    datax.append(locx_2)
    print(locx_2)

    # Data y
    locy_0=re.findall(r"(?<=konumy\: \").*(?=\", baslik:)",lista_2)
    locy_1= (" ".join(locy_0))
    locy_2=(float(locy_1))
    datay.append(locy_2)
    print(locy_2)

    # Region
    region_0=re.findall(r"(?<=baslik\: \").*(?=\", adres:)",lista_2)
    region_1=(" ".join(region_0))
    region.append(region_1)
    print(region_1)

    # Direccion
    dire_0=re.findall(r"(?<=adres\: \").*(?=\", eposta:)",lista_2)
    dire_1=(" ".join(dire_0))
    direccion.append(dire_1)
    print(dire_1)
```

```python
dicc={"Region":nombres,"Latitud":datax,"Longitud":datay,"Address":direccion}
```
