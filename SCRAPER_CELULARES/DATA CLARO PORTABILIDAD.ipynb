```python
# Importo las librerias
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")

driver_path =r'C:\Users\SAD\Downloads\drives_ch\chromedriver.exe'
driver = webdriver.Chrome(driver_path, options=options)
```

```python
# Pagina para celulares prepago - por portabilidad
driver.get('https://tiendaclaro.pe/catalogo/personas/portabilidad-prepago/todos/prepago/todos/todos?nav=1')
sleep(1)
```

```python
Marcas=[]
Modelos=[]
Precio=[]
i=4
vacio="no hay nada"
```

```python
# La caja de donde saco la informacion:
datas =driver.find_elements(By.XPATH,f'//*[@id="contentcatalog"]/app-product/div/div[{i}]/section')

while vacio !=0 :
    
    for data in datas : 

        marca=data.find_element(By.XPATH,f'//*[@id="contentcatalog"]/app-product/div/div[{i}]/section/div[2]/span').text
        Marcas.append(marca)
        print(marca)

        modelo=data.find_element(By.XPATH,f'//*[@id="contentcatalog"]/app-product/div/div[{i}]/section/div[2]/h3').text
        Modelos.append(modelo)
        print(modelo)

        precio=data.find_element(By.XPATH,f'//*[@id="contentcatalog"]/app-product/div/div[{i}]/section/div[3]/div[2]/div[2]/h3/span').text
        Precio.append(precio)
        print(precio)
        
        i+=1
        # para romper
        datas=driver.find_elements(By.XPATH,f'//*[@id="contentcatalog"]/app-product/div/div[{i}]/section')
        vacio=len(datas)

        
        
        



```

```python
dic={"Marca":Marcas ,"Modelo":Modelos , "Precio":Precio}
```

```python
data_Claro_Pre_Portabilidad=pd.DataFrame(dic)
```
