```python
# Importo las librerias
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")

driver_path =r'C:\Users\SAD\Downloads\drives_ch\chromedriver.exe'
driver = webdriver.Chrome(driver_path, options=options)
```

```python
# Pagina para celulares prepago - por portabilidad
driver.get('https://tiendaclaro.pe/catalogo/personas/linea-nueva-prepago/todos/prepago/todos/todos?nav=1')
sleep(1)
```

```python
Marcas=[]
Modelos=[]
Precio=[]
i=1
```

```python
sleep(10)
datas =driver.find_elements_by_xpath('//div[@class="container-product"]')

for data in datas :
    
    try :
        
        marca =data.find_element_by_xpath('.//span[@class="marca"]').text
        Marcas.append(marca)
        print(marca)
    except NoSuchElementException:
        pass
        
    try :
        
        modelo =data.find_element_by_xpath('.//h3[@class="nombre align-items-center d-flex text-center"]').text
        Modelos.append(modelo)
        print(modelo)
    except NoSuchElementException:
        pass
    
        
    try :
        
        precio =data.find_element_by_xpath('.//span[@class="payment-value"]').text
        Precio.append(precio)
        print(precio)
    except NoSuchElementException:
        pass
    
    
    
    
```

```python
dic={"Marcas":Marcas ,"Modelos":Modelos, "Precio":Precio}
```

```python
data_claro_LineaNueva=pd.DataFrame(dic)
```
