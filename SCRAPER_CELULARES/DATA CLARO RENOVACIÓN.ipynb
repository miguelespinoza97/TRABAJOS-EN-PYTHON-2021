```python
# Importo las librerias
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")

driver_path =r'C:\Users\SAD\Downloads\drives_ch\chromedriver.exe'
driver = webdriver.Chrome(driver_path, options=options)
```

```python
# Pagina para celulares prepago - por portabilidad
driver.get('https://tiendaclaro.pe/catalogo/personas/renovacion-prepago/todos/prepago/todos/todos?nav=1')
sleep(1)
```

```python
# Se crea listas donde van estar contenidas todo lo referente a los celulares
Marcas=[]
Modelos=[]
Precio=[]
```

```python
sleep(10)
# Codigo general de donde se extrae la data<
datas =driver.find_elements_by_xpath('//div[@class="container-product"]')
# Imprimos cada elemento haciendo un for 
for data in datas :
    # Ponemos condiciones como el try y except , finalmente un pass en caso de que no encuentre un elemento.
    try :
       # Marca de los celulares  
        marca =data.find_element_by_xpath('.//span[@class="marca"]').text
        Marcas.append(marca)
        print(marca)
    except NoSuchElementException:
        pass
        
    try :
         # Modelo de los celulares
        modelo =data.find_element_by_xpath('.//h3[@class="nombre align-items-center d-flex text-center"]').text
        Modelos.append(modelo)
        print(modelo)
    except NoSuchElementException:
        pass
    
        # Precio de los celulares
    try :
        
        precio =data.find_element_by_xpath('.//span[@class="payment-value"]').text
        Precio.append(precio)
        print(precio)
    except NoSuchElementException:
        pass
    
```

```python
dic={"Marcas":Marcas ,"Modelos":Modelos, "Precio":Precio}
```

```python
data_claro_Renovacion=pd.DataFrame(dic)
```
