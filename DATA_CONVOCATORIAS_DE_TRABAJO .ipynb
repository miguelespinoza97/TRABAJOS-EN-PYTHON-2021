```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")
```

```python
driver_path = r"C:\Users\ANZONY\Downloads\chromedriver.exe"
driver = webdriver.Chrome(driver_path , options =options)
driver.get('https://www.convocatoriascas.com/')
time.sleep(1)
```

```python
i = 1
caja ="vacia"
convocatorias = []
entidades = []
lugares = []
sueldos = []
```

```python
while caja !=0 :
    
    datas = driver.find_elements(By.XPATH,f'/html/body/section[1]/section/article[{i}]')
    
    # Convocatoria
    convocatoria = datas[0].find_element(By.XPATH,f'/html/body/section[1]/section/article[{i}]/div[2]/h2/a').text
    convocatorias.append(convocatoria)
    print(convocatoria)
    
    entidad = datas[0].find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[1]').text
    entidades.append(entidad)
    print(entidad)
    
    lugar=datas[0].find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[2]').text
    lugares.append(lugar)
    print(lugar)
    
    sueldo= datas[0].find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[3]/strong').text
    sueldos.append(sueldo)
    print(sueldo)
    
    i += 1
    
    print( i )
    
    datas = driver.find_elements(By.XPATH, f'/html/body/section[1]/section/article[{i}]')
    caja = len( datas )
                                
    
```
