```python
# Exporto las librerias
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re
from selenium.webdriver.common.alert import Alert
from selenium.common.exceptions import NoSuchElementException
```

```python
# Desabilitamos las extensiones
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')
```

```python
driver_path = "./chromedriver_win32/chromedriver.exe"

driver = webdriver.Chrome(driver_path, options=options)

# Inicializamos el navegador
driver.get('https://gov4c.kz/en/contacts/adresa/631/')
```

```python
adress=[]
nombres=[]
branch=[]
```

```python
caja="no hay"
i=1

while caja!=0 :
    try:

        #Direcciones
        adre=driver.find_element_by_xpath(f'/html/body/main/div[2]/div/div[2]/div/div/article/table/tbody/tr[{i}]/td[4]').text
        adress.append(adre)
        
        print(adre)

        # Nombre de branch
        nombre=driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/h3').text
        branch.append(nombre)
        print(nombre)

        # Nombre completos 
        nombre_2=driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/h3').text
        nom=re.findall(r"(?<=in ).*(?= )",nombre_2) 
        if nom:
            dire=nom[0]
        nombres.append(dire)
        i=i+1
    except:
        caja=0
        print("acabo")
```

```python
dicc={"Adress":adress,"Nombres":nombres,"Branch":branch}
```

```python
registro=pd.DataFrame(dicc)
registro
```
