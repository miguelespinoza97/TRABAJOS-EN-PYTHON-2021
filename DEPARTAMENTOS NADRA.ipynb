```python
# Exporto las librerias
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re
from selenium.webdriver.common.alert import Alert
from selenium.common.exceptions import NoSuchElementException
```

```python
# Desabilitamos las extensiones
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')
```

```python
driver_path = "./chromedriver_win32/chromedriver.exe"

driver = webdriver.Chrome(driver_path, options=options)

# Inicializamos el navegador
driver.get('https://www.nadra.gov.pk/identity/identity-cnic/')
```

```python
# listas donde se guardan los datos
nom_lugar=[]
nom_dau=[]
nom_oic=[]
nom_adre=[]
```

```python
# son 8 elmentos
for j in range (1,9,1):
    # para office Timings
    boton =driver.find_element_by_xpath('//*[@id="menu-item-75724"]').click()
    
    # boton para seleccionar el RH
    rh=driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/div[2]/div/div/div/ul/li[6]/div/div/ul/li[{j}]').click()
    
    sleep(3)
    # paginas 
    pag=driver.find_element_by_xpath('//*[@id="resultsTable_paginate"]').text
    pag_2=pag[-6]
    pag_total=int(pag_2)
    pag_total
  
    for m in range (0,pag_total,1):
        
        

        caja="nada"
        i = 1
        sleep(3)
        while caja!=0 :

            try :

              # Nombre Dau
                dau=driver.find_element_by_xpath(f'//*[@id="historyBody"]/tr[{i}]/td[1]').text
                nom_dau.append(dau)
                print(dau)

                # Nombre oci
                oic=driver.find_element_by_xpath(f'//*[@id="historyBody"]/tr[{i}]/td[2]').text
                nom_oic.append(oic)
                print(oic)

                # Nombre adre
                adre=driver.find_element_by_xpath(f'//*[@id="historyBody"]/tr[{i}]/td[3]').text  
                nom_adre.append(adre)
                print(adre)
                i=i+1

                lugar =driver.find_element_by_xpath('//div[@class="row vertical-align"]').text
                match=re.findall(r"(?<=RHO ).*(?=)",lugar)
                if match:
                    lugar_3=match[0]
                nom_lugar.append(lugar_3)
                print(lugar)

            except:
                caja=0
                print("no hay")
        
        sleep(3)

        driver.find_element_by_xpath('//*[@id="resultsTable_next"]/a').click()

    driver.back()
```

```python
dicc={"LUGAR":nom_lugar,"NOM_DAU":nom_dau,"NOM_OIC":nom_oic,"ADDRESS":nom_adre}
```

```python
registro=pd.DataFrame(dicc)
registro
```
